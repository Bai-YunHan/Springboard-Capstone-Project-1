## Data Wrangling

### Clean the trip data.

Data cleaning procedures conducted in this subsection:
* Handle the missing and duplicated data;
* Find and remove the outliers;
* Modify the format of some data to facilitate the analysis in the next section. 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Concat data from 4 files into one data frame.
# ====================================================================================
# Load files from globing to pandas. 
import glob
csv_files = glob.glob('data/2011*.csv')

list_data = []

for filename in csv_files:
    data = pd.read_csv(filename)
    data.columns = ['duration','start_date','end_date','start_station',\
                    'end_station','bike#','member_type']
    data = data.drop_duplicates()
    data = data.dropna()
    list_data.append(data)

df = pd.concat(list_data, ignore_index=True)

# Use seconds to represent the duration of a trip, with a new column duration_in_sec 
# ====================================================================================
# Reformat the "duration" to "hh:mm:ss" and store in good_time_str
import re
def get_time(time_str):
    matches = re.findall('\d+',time_str)
    return matches[0]+':'+matches[1]+':'+matches[2]

# Conver the good_time_str to integer which is the duration in seconds. 
def get_sec(time_str):
    h, m, s = time_str.split(':')
    return int(h) * 3600 + int(m) * 60 + int(s)

good_time_str = df.duration.apply(get_time)
good_time_str = good_time_str.apply(get_sec)

# Add the series to the DataFrame as column "duration_in_sec"
df['duration_in_sec'] = good_time_str

# Find the outliers.
# ====================================================================================
# Draw a box plot and find the upper wisker.
db = plt.boxplot(df.duration_in_sec)
plt.yscale('log')
plt.ylabel('[seconds]')
plt.xlabel('duration_in_sec')
plt.show()

# Upper_whisker is obtained from the box plot. 
upper_whisker = db['whiskers'][1].get_data()[1][1]

# The lower_whisker is set to 20 instead of 0. 
lower_whisker = 20

print("upper whisker: {:>}s".format(int(upper_whisker)))
print("lower whisker: {:>}s".format(lower_whisker))

# Remove the outliers. 
df = df[df.duration_in_sec < upper_whisker]
df = df[df.duration_in_sec > lower_whisker]

print("outliers are removed.")

# Remove station code from start/end station name. 
# ====================================================================================
# Extracting the name of station. 
def get_name(station_str):
    matches = re.findall('\(\d+\)', station_str)
    code_str = matches[0]
    name_str = station_str.replace(code_str,'')
    return name_str

start_station_name = df.start_station.apply(get_name)
end_station_name = df.end_station.apply(get_name)

# remove the leading and trailing spaces. 
start_station_name = start_station_name.apply(lambda n: n.strip())
end_station_name = end_station_name.apply(lambda n: n.strip())

# add two new columns "start_station_name" and "end_station_name" 
# to store station name without station code. 
df['start_station_name'] = start_station_name
df['end_station_name'] = end_station_name

# Save the clean data to csv file.
# ====================================================================================
# Finally, take the columns with clean data to form the dataframe df_final. 
df_final = df[['start_date',\
               'end_date',\
               'duration_in_sec',\
               'start_station_name',\
               'end_station_name',\
               'bike#',\
               'member_type']]

from datetime import datetime
# save the clean data to csv file. 
# by experiment, using the method below will dramatically 
# reduce the time needed for parsing datetime. 
df_final.to_csv('data/2011-cabi-trip-history-data.csv', index = False)
print("Start parsing datetime data, it will take around 3min to complete.")
# This will take around 3 min!
df_new = pd.read_csv('data/2011-cabi-trip-history-data.csv',parse_dates=[0,1])
print("Done")
# remove the csv file just created. 
import os
os.remove('data/2011-cabi-trip-history-data.csv')
# Reading this file with datetime will only take around 3 sec. 
df_new.to_csv('data/data_generated/2011-cabi-trip-history-data-final.csv', index = False)
```

### Net bike arriving rate of each station in each month

The number of bike getting in and out of the station every month could be used as a metric to measure the importance of each station.

```python
df = pd.read_csv('data/data_generated/2011-cabi-trip-history-data-final.csv', \
                 parse_dates=['start_date','end_date'])
# Making dataframe df_depart: 
# Count the number of departure from each station in every month, store the result in df_depart
# ==============================================================================================
df_depart = df.groupby([df.start_station_name,df.start_date.dt.month]).count()
df_depart.index.names = ['start_station','month']
df_depart.reset_index(inplace=True)
df_depart.drop(['end_date','duration_in_sec','end_station_name','bike#','member_type'], \
               axis=1, inplace=True)
df_depart.rename(columns = {'start_date':'count_depart'},inplace=True)

# Making dataframe df_arr:
# Count the number of arrival to each station in every month, store the result in df_arr
# ==============================================================================================
df_arr = df.groupby([df.end_station_name,df.end_date.dt.month]).count()
df_arr.index.names = ['end_station','month']
df_arr.reset_index(inplace=True)
df_arr.drop(['end_date','duration_in_sec','start_station_name','bike#','member_type'], \
            axis=1, inplace=True)
df_arr.rename(columns = {'start_date':'count_arr'},inplace=True)

# Merge the df_depart and df_arr to dataframe df_combined:
# Bike usage frequency of each station
# ==============================================================================================
df_combined = pd.merge(df_arr, df_depart, left_on = ['end_station','month'], \
                       right_on = ['start_station','month'], how= 'outer')
df_combined = df_combined.fillna(0)

def fill_name(row):
    if row['end_station'] ==0 :
        row['end_station'] = row['start_station']
    if row['start_station'] ==0:
        row['start_station'] = row['end_station']
    return row

df_combined = df_combined.apply(fill_name, axis=1)

df_combined.rename(columns = {'end_station':'station'},inplace=True)
df_combined.drop('start_station',axis=1,inplace=True)
# new column "arr-depart" is the net number of bike arriving the station each month. 
df_combined['arr-depart'] = df_combined.count_arr-df_combined.count_depart
# save the dataframe to a csv file.
df_combined.to_csv('data/data_generated/2011-arr-depart-rate.csv')
```

### The running time of each bike in every month 

```python
df = pd.read_csv('data/data_generated/2011-cabi-trip-history-data-final.csv', \
                 parse_dates=['start_date','end_date'])
df.rename(columns = {'bike#':'bike'},inplace=True)
# find the number of bike 
number_bike = df.bike.unique().shape
print("There are {} bike in this dataset.".format(number_bike[0]))
# find the running time of each bike in every month
df_bike_runtime = df.groupby([df.bike,df.start_date.dt.month])[['duration_in_sec']].sum()
df_bike_runtime.sortlevel(inplace=True)
df_bike_runtime.reset_index(inplace=True)
df_bike_runtime['duration_in_min'] = df_bike_runtime.duration_in_sec/60
df_bike_runtime['duration_in_hour'] = df_bike_runtime.duration_in_sec/3600
df_bike_runtime['duration_in_day'] = df_bike_runtime.duration_in_sec/86400

df_bike_runtime = df_bike_runtime.round(1)
df_bike_runtime.rename(columns = {'start_date':'month'},inplace=True)

# remove the bike with bad code
bad_bike_number = df_bike_runtime.bike
df_bike_runtime = df_bike_runtime[bad_bike_number.str.match('^[W]\d{5}$')]
# save the dataframe to csv file. 
df_bike_runtime.to_csv('data/data_generated/bike_usage_status.csv')
```

### Merge the weather data and trip data

```python
# Generate df_trip_agg
# ===================================================================================================
import pandas as pd

df_trip = pd.read_csv('data/data_generated/2011-cabi-trip-history-data-final.csv',parse_dates=[0,1])

df_trip.index = df_trip.start_date

registered = {'Registered':1,'Casual':0}
casual = {'Registered':0,'Casual':1}
df_trip['cnt_reg'] =df_trip.member_type.map(registered)
df_trip['cnt_cas'] =df_trip.member_type.map(casual)

df_trip_agg = df_trip.resample('H').agg({'start_station_name':'count',\
                                         'cnt_reg':'sum',\
                                         'cnt_cas':'sum',\
                                         'duration_in_sec':'sum'})

df_trip_agg = df_trip_agg.fillna(0)
df_trip_agg.reset_index(inplace=True)
df_trip_agg.rename(columns={'start_date':'date','start_station_name':'count'},inplace=True)
df_trip_agg['duration_in_min'] = df_trip_agg.duration_in_sec/60
df_trip_agg = df_trip_agg.round(0)

# Generate df_cond_final
# ===================================================================================================
df_cond = pd.read_csv('data/hour.csv',parse_dates=[1])
df_cond.rename(columns={'dteday':'date'},inplace=True)
df_cond.index = df_cond.date
df_cond.drop(['instant','date'],axis=1,inplace=True)
# Get the data of 2011
df_cond = df_cond.loc['2011-01-01':'2011-12-31']
df_cond.reset_index(inplace=True)
# Combined to 'date' and 'hr' column to datetime.
df_cond['date'] = df_cond.date + df_cond.hr.astype('timedelta64[h]') 
df_cond_final = df_cond.drop(['season','yr','mnth','hr','weekday',\
                              'workingday','casual','registered','cnt'],axis=1)

# Merge df_trip_agg and df_cond_final
# ===================================================================================================
df_combined = pd.merge(df_trip_agg, df_cond_final, on = 'date', how= 'outer')
df_combined_final = df_combined.dropna()
df_combined_final.to_csv('data/data_generated/2011-trip-and-weather-data.csv', index = False)

df_combined_final.head()
```
